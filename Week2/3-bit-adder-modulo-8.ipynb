Suppose you have two 3 bit numbers $|q_2q_1q_0 \rangle$ and $|q_5q_4q_3 \rangle$, then the in-place adder circuit is as follows

```python
qc_adder = QuantumCircuit(6)
qc_adder.cx(2,5)
qc_adder.ccx(1,4,5)
qc_adder.cx(1,4)
qc_adder.ccx(3,0,1)
qc_adder.ccx(1,4,5)
qc_adder.ccx(3,0,1)
qc_adder.ccx(1,4,5)
qc_adder.ccx(0,3,4)
qc_adder.cx(0,3)
qc_adder.draw()
```

The crux of this problem was to find a sequence that could do a triple and, like a cccx gate, which I was able to do using gates 4 to 7, assuming numbering starts from 1.
